# SciKit Learn Basics

````python
from sklearn import neighbors, datasets, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Iris dataset, a well-known small dataset from scikit-learn
iris = datasets.load_iris()
# Use only the first two features of the iris data for simplicity
X, y = iris.data[:, :2], iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=33)

# Standardize features by removing the mean and scaling to unit variance
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# Create a KNN classifier
knn = neighbors.KNeighborsClassifier(n_neighbors=5)
# Train the classifier
knn.fit(X_train, y_train)
# Predict the labels for the test set
y_pred = knn.predict(X_test)
# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
````

## Understanding the R-squared Evaluation Metric

The R-squared (RÂ²) metric, also known as the coefficient of determination, is a key measure used to evaluate the performance of our regression models. It quantifies the amount of variance in the dependent variable (crop yields) that is predicted from the independent variables (features like weather conditions, pesticide use, etc.).

### How R-squared Works

- **Scale of RÂ²**: R-squared values range from 0 to 1. An RÂ² of 1 indicates that the regression predictions perfectly match the data, an RÂ² of 0 means the model does not explain any of the variance in the target variable around its mean.
- **Interpretation**: The closer the value is to 1, the better the model is at predicting the target variable. An RÂ² close to or below 0 suggests that the model is no better, or worse, than simply predicting the mean of the target variable.

### Model Performance Summary

Our model comparison yields the following R-squared scores for each regression model:
- **Decision Tree Regressor**: The highest RÂ² score at 96%, indicating that this model is highly effective at predicting crop yields, capturing 96% of the variance in the dataset.
- **Gradient Boosting Regressor**: The second highest with an RÂ² score of approximately 90%, also showing strong predictive capabilities.

These scores suggest that both the Decision Tree and Gradient Boosting models are quite capable of capturing the complex patterns in the data, though the Decision Tree Regressor shows slightly superior performance based on this metric.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting dimensions for the plot
a4_dims = (16.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

# Creating the boxplot
sns.boxplot(x="Item", y="hg/ha_yield", palette="vlag", data=yield_df, ax=ax)

# Setting the labels and title
ax.set_xlabel('Crop Type')
ax.set_ylabel('Yield (hg/ha)')
ax.set_title("Distribution of Yields by Crop Type")

# Displaying the plot
plt.show()
````

##Interpreting the Boxplot
In the boxplot:

- **The horizontal line inside each** box represents the median yield for that crop, providing a quick glance at the central tendency.

- **The box itself** encapsulates the interquartile range (IQR), which includes the middle 50% of the data points.
- **The whiskers** extend to show the range of the data, excluding outliers, which are plotted as individual points outside this range.


##Insights from the Plot
This plot can reveal:

- Variability in yield among crops: Crops with wider boxes have more variable yields.
- Outliers: Points outside the whiskers indicate yields that are unusually high or low compared to the typical range for that crop.
- Comparisons of median yields: Directly comparing the medians can indicate which crops generally perform better or worse.

##Conclusion
By visually comparing yield distributions, stakeholders can better understand which crops are most reliable in terms of yield, which are more unpredictable, and where significant outliers may indicate issues or opportunities for further investigation.
